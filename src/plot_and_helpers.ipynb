```python
#jensen shannon div histogrambol
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.spatial.distance import jensenshannon

input_file = "C:\D\Documents\onlab\gretel\\fhrfrequency.csv"
input_file2 ="C:\D\Documents\onlab\signalsrandomvaluefreq.csv"
data = pd.read_csv(input_file)
data2 = pd.read_csv(input_file2)
min_value = data2['FHR'].min()
max_value = data2['FHR'].max()
bin_edges = np.linspace(min_value, max_value, 100 + 1)
hist1, _ = np.histogram(data['FHR'], bins=bin_edges, weights=data['Relative Frequency'], density=True)
hist2, _ = np.histogram(data2['FHR'], bins=bin_edges, weights=data2['Relative Frequency'], density=True)
js_divergence = jensenshannon(hist1, hist2)

print(f"Jensen-Shannon Divergence: {js_divergence}")
```

```python
#jensen shannon divergencia
import numpy as np
from scipy.spatial.distance import jensenshannon

input_file = "C:\D\Documents\onlab\gretel\\fhrfrequency.csv"
input_file2 ="C:\D\Documents\onlab\signalsrandomvaluefreq.csv"
data = pd.read_csv(input_file)
data2 = pd.read_csv(input_file2)

total_frequency = data['Relative Frequency'].sum()
total_frequency2 = data2['Relative Frequency'].sum()
data['Normalized Frequency'] = data['Relative Frequency'] / total_frequency
data2['Normalized Frequency'] = data2['Relative Frequency'] / total_frequency2

merged_data = pd.merge(data, data2, on='FHR', how='outer', suffixes=('_1', '_2'))

merged_data['Normalized Frequency_1'] = merged_data['Normalized Frequency_1'].fillna(0)
merged_data['Normalized Frequency_2'] = merged_data['Normalized Frequency_2'].fillna(0)

assert np.isclose(merged_data['Normalized Frequency_1'].sum(), 1), "data does not sum to 1"
assert np.isclose(merged_data['Normalized Frequency_2'].sum(), 1), "data2 does not sum to 1"

js_divergence = jensenshannon(merged_data['Normalized Frequency_1'], merged_data['Normalized Frequency_2'])

print(f"Jensen-Shannon Divergence: {js_divergence}")
```

```python
#Random fajl kivalasztas kisebb adatbazishoz
import os
import random
import shutil
import pandas as pd

def select_and_copy_files(source_folder, destination_folder, max_rows, num_files):
 
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    eligible_files = []

    for file_name in os.listdir(source_folder):
        if file_name.endswith('.csv'):
            file_path = os.path.join(source_folder, file_name)
          
            with open(file_path, 'r') as file:
                row_count = sum(1 for line in file)
        
            if row_count <= max_rows:
                eligible_files.append(file_path)

    
    selected_files = random.sample(eligible_files, min(num_files, len(eligible_files)))

    
    for file_path in selected_files:
        file_name = os.path.basename(file_path)
        shutil.copy(file_path, os.path.join(destination_folder, file_name))
        print(f"File '{file_name}' copied to '{destination_folder}'")

source_folder = "C:\D\Documents\onlab\signals_noseconds"
destination_folder = "C:\D\Documents\onlab\signalsrandom"
max_rows = 700  # max sorok szama
num_files = 552  # hany fajlt akarunk

select_and_copy_files(source_folder, destination_folder, max_rows, num_files)
```

```python
#hosszfrequency
import os
import csv

def count_csv_lengths(folder_path):
    lengths_frequency = {}  # Dictionary to store frequency of lengths
    
    # Iterate over files in the folder
    for file_name in os.listdir(folder_path):
        if file_name.endswith(".csv"):
            file_path = os.path.join(folder_path, file_name)
            with open(file_path, 'r') as file:
                num_lines = sum(1 for line in file)  # Count number of lines in the file
                lengths_frequency.setdefault(num_lines, 0)
                lengths_frequency[num_lines] += 1
                
    return lengths_frequency

def save_frequency_to_csv(frequency, output_file):
    with open(output_file, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['Length', 'Frequency'])
        for length, count in frequency.items():
            writer.writerow([length, count])

# Example usage
folder_path = "C:\D\Documents\onlab\signalsrandom"
output_file = "C:\D\Documents\onlab\signalsrandomlengthfrequency.csv"

frequency = count_csv_lengths(folder_path)
save_frequency_to_csv(frequency, output_file)
print(f"Frequency of CSV file lengths saved to {output_file}")
```
